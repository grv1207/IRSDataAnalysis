```python
def extract(string):
    revenue = {}
    urldata = url.request.urlopen('https://s3.amazonaws.com/irs-form-990/'+string).read()  
    tree = ET.fromstring(urldata)
    try:
        
        for neighbor in tree.iter(tag='{http://www.irs.gov/efile}TotalRevenuePriorYear'):
            #print(neighbor.text)
            revenue['TotalRevenuePriorYear'] = neighbor.text
            #revenue.append(neighbor.text)
        for neighbor in tree.iter(tag='{http://www.irs.gov/efile}TotalRevenueCurrentYear'):
            #print(neighbor.text)
            revenue['TotalRevenueCurrentYear']= neighbor.text
        for neighbor in tree.iter(tag='{http://www.irs.gov/efile}TaxYear'):
            #print(neighbor.text)
            revenue['TaxYear'] = neighbor.text
        for neighbor in tree.iter(tag='{http://www.irs.gov/efile}State'):
            #print(neighbor.text)
            revenue['State'] = neighbor.text
            
        return revenue   
    except:
        return 'no values'
    
```
